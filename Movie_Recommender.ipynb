# Movie Recommender System
Import the useful libraries required for the model.

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(10)
```

```python
credits.head(10)
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head()
```

```python
print(movies.columns)
```

```python
movies['crew'][0]
```

```python
movies = movies[['title', 'movie_id', 'genres', 'keywords', 'overview', 'cast', 'crew']]
```

```python
movies.isnull().value_counts()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().value_counts()
```

```python
movies.duplicated().value_counts()
```

```python
movies.duplicated().sum()
```

```python
import ast
```

```python
def convert(object):
    l = []
    for i in ast.literal_eval(object):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_c(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_c)
```

```python
def director(obj):
    l =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(' ', '') for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(' ', '') for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(' ', '') for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(' ', '') for i in x])
```

```python
movies.head(30)
```

```python
movies['tag'] = movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

#### Let's create a new dataset after preprocessing, which will include the movie id, title and tag.

```python
new_df = movies[['movie_id', 'title', 'tag']]
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
new_df['tag'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])
```

## This is our main function...!ðŸ˜€

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1 : 6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

## And here we go.....ðŸ™‚

```python
recommend('Harry Potter and the Half-Blood Prince')
```

```python
recommend('Captain America: Civil War')
```

```python
recommend('Robin Hood')
```
